```python
from tensorflow.keras import models
from tensorflow.keras import layers

First_model = models.Sequential()

First_model.add(layers.Dense(512 , activation = 'relu', input_shape = (28 * 28,)))
First_model.add(layers.Dense(10, activation = 'softmax'))
```

```python
First_model.compile(optimzer="rmsprop", loss = "categorical_crossentropy", metrices = ['accuracy']  )
```

```python
type(train_images)
```

```python
train_images = train_images.reshape(60000,28*28)
train_images = train_images.astype("float32")/255
test_images = test_images.reshape(10000,28*28)
test_images = test_images.astype("float32")/255
```

```python
from tensorflow.keras.utils import to_categorical

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
First_model.fit(train_images, train_labels, epochs = 25 , batch_size = 128)
```

```python
test_loss, test_acc = First_model.evaluate(test_images, test_labels)
```
